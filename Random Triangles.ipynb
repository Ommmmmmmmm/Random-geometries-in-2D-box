```python
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
l, b, h = 95, 95, 0
d = 7
points = 100
```

```python
max_points = math.floor((l/d)*(b/d)*(1 if h == 0 else (h/d)))
if(max_points <= points):
    print("Cannot")
    die()
  
lop = []
print("Max possible points: ", max_points)
print("Requested points: ", points)
def test_point(x, y, z):
  for p in lop:
    if(np.sqrt(np.power(p[0] - x, 2) + np.power(p[1] - y, 2) + (0 if h == 0 else np.power(p[2] - z, 2))) < d):
      return False
  return True
```

```python
while True:
  x = np.random.rand()*l
  y = np.random.rand()*b
  z = np.random.rand()*h
  
  if(test_point(x, y, z)):
    lop.append([x, y, z])

  
  if(len(lop) == points):
    break;
```

```python
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(*zip(*lop))
print(len(lop), "points")
plt.show()
```
